#### 2. Import setup (numpy, pandas, pyplot, seabor)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

#### 3. Dataframe specification

```python
#sample seaborn dataset
print(sns.get_dataset_names())
```

```python
crash_df= sns.load_dataset('car_crashes')
crash_df.head()
```

```python
crash_df.info()
```

## Distriburion Plots

#### (i). Distribution Plot

```python
sns.distplot(crash_df['not_distracted'], hist=True, kde=False, bins = 25)
```

#### (ii). Joint Plots

```python
sns.jointplot(x='speeding', y= 'alcohol', data = crash_df, kind ='reg')
```

#### (iii). KDE Plots (Kernel Density Distribution Data)

```python
sns.kdeplot(crash_df['alcohol'])
```

#### (iv). Pair plots

```python
sns.pairplot(crash_df)
```

```python
#pair plot for categorical data

tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df[['total_bill', 'sex']], hue ='sex', palette='Blues')
```

#### (V). Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

### Sytling
Making our plots look nicer by adding(colur, size)

```python
sns.set_style('ticks')
plt.figure(figsize=(8,14))

sns.set_context('paper', font_scale =1.4)
sns.jointplot(x='speeding', y='alcohol', data = crash_df, kind='reg')
```

## Categorical Plot
#### Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data = tips_df, estimator=np.mean, palette='hsv')
```

#### Count Plot

```python
sns.countplot(x='sex', data=tips_df, palette='hsv')
```

```python
tips_df.head()
```

#### Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='hsv')
plt.legend(loc=0)
```

#### Violin Plot

```python
sns.violinplot(data=tips_df, x='day', y='total_bill', palette='hsv')
```

### Swarm Plot

**Palette:** It determines the colour map: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x='day', y='total_bill', data=tips_df, palette='hsv')
sns.set_style('dark')
sns.set_context('talk')
```
